## Beispiel

```python
# Import von Bibliotheken
import numpy as np
import matplotlib.pyplot as plt

# Berechnung
np.random.seed(0)
n = 100000
x = np.random.standard_normal(n)
y = 2.0 + 3.0 * x + 4.0 * np.random.standard_normal(n)
xmin = x.min()
xmax = x.max()
ymin = y.min()
ymax = y.max()

# Ausgabe
fig, ax = plt.subplots(figsize=(7,4))
hb = ax.hexbin(x, y, gridsize=50, bins='log', cmap='inferno')
ax.axis([xmin, xmax, ymin, ymax])
ax.set_title("log color scale")
cb = fig.colorbar(hb, ax=ax)
cb.set_label('log10(N)')
plt.show()
```

## Slicing – Extrem hilfreich

- Slicing ermöglicht Zugriff auf viele Elemente 
- Slicing dezimiert die Elemente nach gewünschtem Muster
<br><br>
- Syntax:   *start:end:step*
- Defaults
 - *start*: 0
 - *end*: letztes Element
 - *step*: 1
<br><br>
- *end* wird nicht angenommen
- bei negativem *start* oder *end* wird vom letzten Element aus gezählt

### Einfache Sequenz

```python
# Beispiel
txt = '0123456789'
slice = txt[6:7] # beginnend ab dem 2. Element bis zum 4. Element in Einerschritten
print(slice)
```

```python
slice_2 = txt[-5::1] # vom 5.-letzten Element bis zum Ende in Einerschritten
slice_2
```

## if / elif / else 

- Blöcke durch Einrücken (Tab), **KEINE** Klammern!
- Pythonic: Tab von 4 Zeichen
- Bedingungs-Statement endet mit *:*


### Syntax

- if boolscher Ausdruck:
    - statements
- elif boolscher Ausdruck:
    - statements
- else:  
    - statements

### Beispiel

```python
import time

if (np.random.rand() > 1/3):
    print('...')
    time.sleep(1)
    print('Hä?')
elif np.random.rand() > 0.5:
    print('wie bitte')
else:
    print('alles klar!')
```

- call-by-reference

- Objekte werden zurückgegeben, neu zugewiesene Objekte nicht!

## Keyword Parameter, Defaults

### Parameterarten
- Positions-parameter (klassisch)
- Keyword Parameter: keyw=default

```python
import numpy as np
def zylVolumen(r=0.1, h=1):
    return np.pi*r*r*h
```

```python
# Können weggelassen werden
zylVolumen()
```

```python
# Können einzeln übergeben werden (erster Parameter)
zylVolumen(2)
```

```python
# Identifizierbar über Keyword
zylVolumen(r=0.2)
```

```python
# Können einzeln übergeben werden (ab dem zweiten Parameter)
zylVolumen(h=2)
```

```python
# Identifizierbar über Position
zylVolumen(0.5, 0.5)
```

```python
# in beliebiger Reihnfolge
zylVolumen(h=1 , r=0.2)
```

### Beliebig lange Parameter-Listen
- häufig in Bibliotheken
- Positionsparameter: \*arg
- Keyword Parameter: \*\*kwargs

```python
def varParam(*args, **kwargs):
    print('Positionsargumente:')
    for arg in args:
        print(arg)
    print('Keyword Argumente:')
    for kwarg in kwargs:
        print(kwarg, ':', kwargs[kwarg])
            
```

```python
varParam('Test', 42, key='keyval', h=5)
```
